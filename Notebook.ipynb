```python
import pandas as pd
# Daten aus der CSV-Datei lesen
data = pd.read_csv("CAERS_ASCII_2004_2017Q2.csv")
```

```python
import matplotlib.pyplot as plt

#Teil von Klaus
#Frage: What are the most commonly reported foodstuffs?

# Nur die Zeilen behalten, in denen das gemeldete Produkt ein Lebensmittel ist
food_data = data[data['PRI_FDA Industry Name'].str.contains('Food', case=False, na=False)]

# Lebensmittel und deren Hufigkeit zhlen
food_counts = food_data['PRI_Reported Brand/Product Name'].value_counts()

# Die Top 10 am hufigsten gemeldeten Lebensmittel auswhlen
top_food_counts = food_counts.head(10)

# Balkendiagramm erstellen
plt.figure(figsize=(10, 6))
top_food_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 am hufigsten gemeldeten Lebensmittel')
plt.xlabel('Lebensmittel')
plt.ylabel('Anzahl der Berichte')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
